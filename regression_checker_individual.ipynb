```python

!pip install pandas matplotlib seaborn

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from google.colab import files

uploaded = files.upload()
file_name = list(uploaded.keys())[0]
data = pd.read_excel(file_name)


X = data[['Empirical h* Value']].values.reshape(-1, 1)
y = data['Approx h* Value'].values


model = LinearRegression(fit_intercept=False)
model.fit(X, y)
slope = model.coef_[0]
y_pred = model.predict(X)
r2 = r2_score(y, y_pred)
plt.figure(figsize=(10, 6))
sns.scatterplot(x=X.flatten(), y=y, label='Data points')
plt.plot(X, y_pred, color='red', label=f'Regression line: y = {slope:.2f}x\nR = {r2:.2f}')
plt.xlabel('Empirical h* Value')
plt.ylabel('Approx h* Value')
plt.title('Regression Plot Constrained to Origin')
plt.legend()
plt.grid(True)
plt.show()

print(f"Slope (Gradient): {slope}")
print(f"R value: {r2}")
```
